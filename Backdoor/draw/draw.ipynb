```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total = np.load('./singleround_iid/Semantic-backdoorsaccuracy.npy')
normal = np.load('./singleround_iid/Semantic-backdoorsaccuracy_green_car.npy')
backdoor = np.load('./singleround_iid/Semantic-backdoorsaccuracy_green_car_backdoor.npy')
plt.plot(total, label='total')
plt.plot(normal, label='normal')
plt.plot(backdoor, label='backdoor')
plt.title('singleround_iid_Semantic')
plt.xlabel('1')
plt.ylabel('2')
plt.legend()
plt.show()
```
